### Indexing and extracting data


 **What is the share of clients attracted in our DataFrame?**


```python
print("Share of attracted clients =", '{:.1%}'.format(df["y"].mean()))
```

10.9% is a rather bad indicator for a bank, with such a percentage of attracted customers a business can collapse.


 **What are the mean values ​​of numerical features among the attracted clients?**

```python
df["y"] = df["y"].replace({"no": 0, "yes": 1})

# Take mean only over numeric columns
mean_attracted = df[df["y"] == 1].mean(numeric_only=True)
print(mean_attracted)
```

Thus, the average age of the attracted clients is about 40 (`age` = 40.89), and 2 calls were required to attract them (`campaign` = 2.15).


 **What is the average call duration for the attracted clients**?


```python
acd = round(df[df["y"] == 1]["duration"].mean(), 2)
acd_in_min = acd // 60
print("Average call duration for attracted clients =", acd_in_min, "min", int(acd) % 60, "sec")
```

So, the average duration of a successful call 9.0 min 20 sec.


**What is the average age of attracted (`y == 1`) and unmarried (`'marital' == 'single'`) clients?**


```python
print("Average age of attracted clients =", int(df[(df["y"] == 1) & (df["marital"] == "single")]["age"].mean()), "years")
```

The average age of unmarried attracted clients is 32, which should be considered when working with such clients.


## Pivot tables


Ler's see how observations in our sample are distributed in the context of two features - `y` and `job`. 

```python
pd.crosstab(df["y"],
            df["job"],
            normalize = 'index')
```

People in `administrative job (29%)` tend to attract more followed by people in `technician(18%)` and `blue collar(14%)` job.

Two features - `y` and `marital`

```python
pd.crosstab(df["y"],
            df["marital"],
            normalize = 'index')
```

We can see that more than half of the clients (62%, column `married`) are married and have not issued a deposit.


Let's find the average age and the call duration for different types of client employment `job`:


```python
df.pivot_table(
    ["age", "duration"],
    ["job"],
    aggfunc = "mean",
).head(10)
```

The obtained results allow us to plan marketing banking campaigns more effectively.


## Visualization in Pandas


Creating a pairwise scatter plot of `age`,`duration` and `campaign` features

```python
pd.plotting.scatter_matrix(
    df[["age", "duration", "campaign"]],
    figsize = (15, 15),
    diagonal = "kde")
plt.show()
```

The distribution is similar to normal for age, and for a call duration and the number of contacts, the [geometric distribution](https://en.wikipedia.org/wiki/geometric_distribution?utm_medium=Exinfluencer\&utm_source=Exinfluencer\&utm_content=000026UJ\&utm_term=10006555\&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkQuickLabsEDA_Pandas_Banking_L126457256-2021-01-01) is more suitable.


**Histogram for `age` feature:**


```python
df["age"].hist()
```

The histogram shows that most of our clients are between the ages of 25 and 50, which corresponds to the actively working part of the population.


**Histogram for features all together:**


```python
df.hist(color = "k",
        bins = 30,
        figsize = (15, 10))
plt.show()
```

Using boxplot


```python
df.boxplot(column = "age",
           by = "marital")
plt.show()
```

The plot shows that unmarried people are on average younger than divorced and married ones. For the last two groups, there is an outlier zone over 70 years old, and for unmarried - over 50.


**Now we will try to do this by data grouping on other features:**

```python
df.boxplot(column = "age",
           by = ["marital", "housing"],
           figsize = (20, 20))
plt.show()
```

As we can see, age and marital status do not have any significant influence on having a housing loan.
